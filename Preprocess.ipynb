```python
import pandas as pd
import pathlib as pl
import xml.etree.ElementTree as ET
import cv2
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Get the names from the xml files and add them to a dictionary
names = []
for i in range(877):
    xmlFile = pl.Path(f'./annotations/road{i}.xml')
    # Add the name value from the xml file to the dictionary for the key 'road' + i
    tree = ET.parse(xmlFile)
    root = tree.getroot()
    name = root.find('object').find('name').text
    names.append(name)
```

```python
# Import the first image located in the images folder
imgPath = pl.Path('./images/road0.png')
img = cv2.imread(imgPath)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# display the image
plt.imshow(img)
```

```python
images = []
for i in range(877):
    imgFile = pl.Path(f'./images/road{i}.png')
    img = cv2.imread(imgFile)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    images.append(img)
```

```python
import pickle
with open('images.pkl', 'wb') as f:
    pickle.dump(images, f)
```

```python
with open('target.pkl', 'wb') as f:
    pickle.dump(names, f)
```
